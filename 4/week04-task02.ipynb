2- На загруженных данных обучите преобразование PCA с числом компоненты равным 10. Скольких компонент хватит, чтобы объяснить 90% дисперсии?

2- On the downloaded data, teach the PCA conversion with a component number of 10. How many components are sufficient to account for 90% of the variance?


```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca.fit(data)

var=pca.explained_variance_ratio_
kans=0
for k in range(10):
    if sum(var[:k])>=0.9:
        kans=k
        break

from lib.helpersmd import write_to_file
write_to_file(kans,"week%d-A%dT%d.out"%(4,2,1))
```
