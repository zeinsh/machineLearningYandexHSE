```python
import numpy as np
import pandas
from lib.helpersmd import write_to_file
```

1- load the training and test samples from the files perceptron-train.csv and perceptron-test.csv. The target variable is written in the first column, the features are in the second and third.

1- Загрузите обучающую и тестовую выборки из файлов perceptron-train.csv и perceptron-test.csv. Целевая переменная записана в первом столбце, признаки — во втором и третьем.

```python
#1 load dataset , split X,Y
train=pandas.read_csv('perceptron-train.csv',header=None)
train=train.as_matrix(columns=None)
ytrain=train[:,0]
Xtrain=train[:,1:]

#load testset , split X,Y
test=pandas.read_csv('perceptron-test.csv',header=None)
test=test.as_matrix(columns=None)
ytest=test[:,0]
Xtest=test[:,1:]
```

```python
# Plot also the training points
from lib.plotmd import scatterData
import matplotlib.pyplot as plt
classmap={-1:"Negatives",1:"Positives"}
scatterData(Xtrain,ytrain,classmap)
plt.title("X1 and X2 are on different scales")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.show()
```

2- Train the perceptron with the standard parameters and random_state = 241.

2- Обучите персептрон со стандартными параметрами и random_state=241.

```python
from sklearn.linear_model import Perceptron
random_state=241
clf = Perceptron(random_state=random_state,verbose=True)
clf.fit(Xtrain, ytrain)
```

3- Calculate the quality (the proportion of correctly classified objects, accuracy) of the obtained classifier on the test sample.

3- Подсчитайте качество (долю правильно классифицированных объектов, accuracy) полученного классификатора на тестовой выборке.

```python
predictions = clf.predict(Xtest)
quality_without_norm=(predictions==ytest).mean()
```

```python
import matplotlib.pyplot as plt
from lib.plotmd import scatterData , plotDecisionBoundary

truepositive=np.multiply((predictions==1),(ytest==1))
truenegative=np.multiply((predictions==-1),(ytest==-1))*2
falsepositive=np.multiply((predictions==1),(ytest==-1))*3
falsenegative=np.multiply((predictions==-1),(ytest==1))*4
yscatter=truepositive+truenegative+falsepositive+falsenegative
classmap={1:"True Positive",
          2:"True Negative",
          3:"False Positive",
          4:"False Negative"}
scatterData(Xtest,yscatter,classmap)

fig_boundary={}
fig_boundary["x_min"],fig_boundary["x_max"]= min(Xtest[:,0]),max(Xtest[:,0])
fig_boundary["y_min"],fig_boundary["y_max"]= min(Xtest[:,1]),max(Xtest[:,1])
w=clf.coef_[0]
b=clf.intercept_
plotDecisionBoundary(w,b,fig_boundary)

plt.title("predictions")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.show()
```

4- Normalize the training and test sample using the StandardScaler class.

4- Нормализуйте обучающую и тестовую выборку с помощью класса StandardScaler.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Xtrain_scaled = scaler.fit_transform(Xtrain)
Xtest_scaled = scaler.transform(Xtest)
```

```python
# Plot also the training points after scaling
from lib.plotmd import scatterData
import matplotlib.pyplot as plt
classmap={-1:"Negatives",1:"Positives"}
scatterData(Xtrain_scaled,ytrain,classmap)
plt.title("X1 and X2 after scaling")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.show()
```

5- Train the perceptron on a new sample. Find the proportion of correct answers to the test sample.

5- Обучите персептрон на новой выборке. Найдите долю правильных ответов на тестовой выборке.

```python
from sklearn.linear_model import Perceptron
random_state=241
clf = Perceptron(random_state=random_state)
clf.fit(Xtrain_scaled, ytrain)
predictions = clf.predict(Xtest_scaled)
quality_with_norm=(predictions==ytest).mean()
```

```python
def getLine1(w,bias,fig_boundary):
    import numpy as np
    x_min,x_max=fig_boundary["x_min"],fig_boundary["x_max"]
    y_min,y_max=fig_boundary["y_min"],fig_boundary["y_max"]
    print w
    if w[1]!=0 and abs(w[0])<abs(w[1]):
        yy=np.array([y_min,y_max])
        xx=(-bias-yy*w[1])/w[0]
    else:
        xx=np.array([x_min,x_max])
        yy=(-bias-yy*w[0])/w[1]
    return xx,yy
```

```python
import matplotlib.pyplot as plt
from lib.plotmd import scatterData,plotDecisionBoundary

truepositive=np.multiply((predictions==1),(ytest==1))
truenegative=np.multiply((predictions==-1),(ytest==-1))*2
falsepositive=np.multiply((predictions==1),(ytest==-1))*3
falsenegative=np.multiply((predictions==-1),(ytest==1))*4
yscatter=truepositive+truenegative+falsepositive+falsenegative
classmap={1:"True Positive",
          2:"True Negative",
          3:"False Positive",
          4:"False Negative"}
scatterData(Xtest_scaled,yscatter,classmap)

fig_boundary={}
fig_boundary["x_min"],fig_boundary["x_max"]= min(Xtest_scaled[:,0]),max(Xtest_scaled[:,0])
fig_boundary["y_min"],fig_boundary["y_max"]= min(Xtest_scaled[:,1]),max(Xtest_scaled[:,1])
w=clf.coef_[0]
b=clf.intercept_
plotDecisionBoundary(w,b,fig_boundary)

plt.title("predictions")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.show()
```

6- Find the difference between the quality on the test sample after the normalization and the quality before it. This number is the answer to the task.

6- Найдите разность между качеством на тестовой выборке после нормализации и качеством до нее. Это число и будет ответом на задание.

```python
quality_diff=quality_with_norm-quality_without_norm
write_to_file(quality_diff,"week%d-A%dT%d.out"%(2,3,1))
```
